# A function that takes input search tag and out put with a csv file an head of 10 row



```python


def tweetSearch(topic):
    public_tweets = api.search(topic)
    
    tweetsOnTopic = []
    polarity = []
    subjectivity = []
    sentimentType = []
    
    for tweet in public_tweets:
        tweetsOnTopic.append(tweet.text)
        #Step 4 Perform Sentiment Analysis on Tweets
        analysis = TextBlob(tweet.text)

        polarity.append(analysis.sentiment.polarity)
        subjectivity.append(analysis.sentiment.subjectivity)
        if (analysis.sentiment.polarity > 0):
            sentimentType.append("positive")
        elif(analysis.sentiment.polarity < 0):
            sentimentType.append("negative") 
        elif(analysis.sentiment.polarity == 0):
            sentimentType.append("neutral")

        df = pd.DataFrame(data={"tweetsOn{}".format(topic): tweetsOnTopic, "polarity": polarity, "subjectivity": subjectivity, "sentimentType": sentimentType})
        df.to_csv("twitter_{}_tweets.csv".format(topic), sep=',',index=False)
        
    return df.head(10)

#Try With other names

# tweetSearch('Data Science')
tweetSearch('India')
```
